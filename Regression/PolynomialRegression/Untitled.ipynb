```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:, 1:2].values # 1:2 so X is a matrix
y = dataset.iloc[:, 2].values
```

```python
# Linear Regressuin
from sklearn.linear_model import LinearRegression
Lregressor = LinearRegression()
Lregressor.fit(X, y)
```

```python
# Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures 
# Transforms X to have its subssequent powers (X², X³, etc)
Pregressor = PolynomialFeatures(degree =4)
X_poly = Pregressor.fit_transform(X)
PLregressor = LinearRegression()
PLregressor.fit(X_poly, y)
```

```python
# Visualizing Linear Regressions results
plt.scatter(X, y, color = 'red')
plt.title("Linear Regression Results")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.plot(X, Lregressor.predict(X), color='blue') 
```

```python
# Visualizing Polynomial Regressions results
X_grid = np.arange(min(X), max(X), 0.1) # vector
X_grid = X_grid.reshape((len(X_grid), 1)) # matrix
plt.scatter(X, y, color='red')
plt.plot(X_grid, PLregressor.predict(Pregressor.fit_transform(X_grid)), color='blue')
plt.title("Polynominal Regression results")
plt.xlabel("Position Level")
plt.ylabel("Salary")
```
